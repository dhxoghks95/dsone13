# 상영시간 통합

```python
def min_to_(x):
    if pd.notnull(x):
        return float(x[:-1])
```

```python
def hour_to_(x):
    if pd.notnull(x):
        if '시간' in x:
            return int(x[0])*60 + float(x[4:-1])
        else:
            return float(x[:-1])
```

```python
DT.씨네21_상영시간 = DT.씨네21_상영시간.apply(min_to_)
DT.왓챠_상영시간 = DT.왓챠_상영시간.apply(hour_to_)
```

```python
DT.씨네21_상영시간.fillna(DT.왓챠_상영시간, inplace=True)
```

```python
DT = DT.drop('왓챠_상영시간', axis=1)
```

# 연도 통합

```python
DT.씨네21_연도.fillna(DT.왓챠_연도, inplace=True)
DT.씨네21_연도.fillna(DT.다음_연도, inplace=True)
```

```python
DT = DT.drop(['왓챠_연도', '다음_연도'], axis=1)
```
